```python
from IPython.display import clear_output

def clear():
    clear_output(wait=True)


class DrugStore:

  list_drugStore = {
      "shampoo_dove": {
          "price": 12.90,
          "amount": 400,
          "lot": "SH1234A",
      },
      "condicionador_dove": {
          "price": 13.50,
          "amount": 400,
          "lot": "CD5678B",
      },
      "sabonete_nivea": {
          "price": 2.99,
          "amount": 90,
          "lot": "SB4321C",
      },
      "creme_dental_colgate": {
          "price": 4.79,
          "amount": 90,
          "lot": "CD9988D",
      },
      "escova_dental_oralb": {
          "price": 6.99,
          "amount": 1,
          "lot": "ED7766E",
      }
  }

  # Start the DrugStore class
  def __init__(self, list_drugStore):
    self.list_drugStore = list_drugStore

  # Method to show products in {list_drugStore}
  def showProducts(self):
    for product in self.list_drugStore:
      print(f"{product.replace('_', ' ')} | R$ {self.list_drugStore[product]['price']} | Qty: {self.list_drugStore[product]['amount']} | Lot: {self.list_drugStore[product]['lot']}")

  # Method to add a new product to {list_drugStore}
  def newProduct(self):
    while True:

        # Get information
        nameProduct = input('Product name: ').replace(' ', '_').lower()
        priceProduct = float(input('Product price: '))
        amountProduct = int(input('Product amount in stock: '))
        lotProduct = input('Product lot: ')

        print('Are you sure you want to add this product? (Y/N): ')
        quest = input('> ').upper()

        # If the information is correct:
        if quest == 'Y':
            # Product added
            self.list_drugStore[nameProduct] = {
                "price": priceProduct,
                "amount": amountProduct,
                "lot": lotProduct,
            }
            print("Added!")
        else:
            print("Canceled!")

        print("Add another? (Y/N): ")
        sair = input('> ').upper()

        # Close method
        if sair == 'N':
            break

  # Method to delete a specific product
  def delProduct(self):
    while True:

        # Show products in dict
        for product in self.list_drugStore:
            print(product.replace('_', ' '))

        print('\nProduct name (or ESC to exit): ')
        deletar = input('> ').replace(' ', '_').lower()

        # Close method
        if deletar == 'esc':
            break

        if deletar not in self.list_drugStore:
            print('Product not found!')
        else:
            print(f"Are you sure you want to delete {deletar}? (Y/N)")
            quest = input('> ').upper()

            if quest == 'Y':
                del self.list_drugStore[deletar]
                print('Product deleted!')
                input("Press Enter to continue...")
                break
            else:
              print(f"{deletar} was not deleted.")
              input("Press Enter to continue...")
              break

  # Menu method
  def show_opptions(self):
    op = input('---------- choose an option: ----------\n- 1 | Add new product\n- 2 | Delete a product\n- 3 | Show Products\n\n0) Exit\n')

    match op:
        case '1':
            self.newProduct()
        case '2':
            self.delProduct()
        case '3':
            self.showProducts()
            input("\nPress Enter to continue...")
        case '0':
            print("Out...")
            exit()

  # Method to start the system
  def start(self):
    while True:
      clear()
      self.show_opptions()


# ---------------------------------------------------- #
compras_drugStore = DrugStore(DrugStore.list_drugStore)
compras_drugStore.start()
```
